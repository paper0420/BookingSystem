```python
# Write the Python code for Task 1 here
account_file = "account.txt"
def init_system(account_file):
    try:
        print("Welcome to Online Driver Knowledge Test (DKT) Booking") 
        file = open(account_file, 'r')
        if(file):
            print("The account file is existing: Please Log In")
        file.close()
        return True
    except FileNotFoundError:
        print("The account file is unavailable: Please Sign Up")
        return False
        
def create_file(account_detail):
    file = open("account.txt",'w')
    for key,value in account_detail.items():
        file.write(f"{key},{value}\n")
    file.close()

def read_file(account_file):
    account_detail = {}
    file = open(account_file,'r')
    for line in file:
        key,value = line.split(',')
        account_detail[key] = value.strip()
    file.close()
    return account_detail
    

def log_in(account_file):

    account_detail = read_file(account_file)
   
    print("----Log In----\n")
    is_login_pending = True
    while(is_login_pending):
        id = input("Enter Account ID: ")
        password = input("Enter password")
        if(id == account_detail['id'] and password == account_detail['password']):
            return account_detail
        elif id != account_detail['id']:
            print("Invalid Account ID")
            continue
        elif password != account_detail['password']:
            print("Incorrect Password. To reset, please enter your registered mobile number.")
            mobile = input("Enter mobile number : ")
            if(mobile == account_detail['mobile']):
                password = input("Reset password (The password must contain 6 characters) : ")
                if len(password) == 6:
                    account_detail['password'] = password
                    create_file(account_detail)
                    account_detail = read_file(account_file)
                    return account_detail
                else:
                    print("The password must contain 6 characters")
                
            else:
                print("Your information is not correct.")
                return False
           
def sign_up():
    print("----Sign Up----\n")
    
    account_detail = {
        'id': 0,
        'mobile':0,
        'password':''
    }
    
    is_id_pending = True
    while(is_id_pending):
        photoid = input("Enter photo ID number (PhotoID number must contain 10 digits) : ")
        if (photoid.isdigit() and len(photoid) == 10):
            account_detail['id'] = (int)(photoid)
            is_id_pending = False
        else:
            print("Photo ID is invalid")

    is_mobile_pending = True
    while(is_mobile_pending):
        mobile = input("Enter mobile number (Mobile number must contain 10 digits) : ")
        if (mobile.isdigit() and len(mobile)==10):
            account_detail['mobile'] = (int)(mobile)
            is_mobile_pending = False
        else:
            print("Mobile number is invalid")

    is_pass_pending = True
    while(is_pass_pending):
        password = input("Set password (The password must contain 6 characters) :")
        if len(password) == 6:
            account_detail['password'] = password
            is_pass_pending = False
        else:
            print("The password must contain 6 characters")
    
    create_file(account_detail)
    print("** Sign up completed ** Please Log in **")

def show_menu(option):
    if(option == 1):
        print("Main Menu")
        print("-"*10+"Press 1 "+"-"*10 +" Show Main Menu")
        print("-"*10+"Press 2 "+"-"*10 +" Check your booking status")
        print("-"*10+"Press 3 "+"-"*10 +" Book online DKT")
        print("-"*10+"Press 4 "+"-"*10 +" Reschedule your DKT")
        print("-"*10+"Press 5 "+"-"*10 +" Cancel your DKT")
        print("-"*10+"Press 0 "+"-"*10 +" Logout")
        print("-"*10+"Press 00"+"-"*10 +" Exit the program.")
    elif option == 2:
        print("-"*10+"Press 1 "+"-"*10 +" Show Main Menu")
        print("-"*10+"Press 0 "+"-"*10 +" Logout")
        print("-"*10+"Press 00"+"-"*10 +" Exit the program.")
    elif option ==3:
        print("-"*10+"Press 9 "+"-"*10 +" Login")
        print("-"*10+"Press 00"+"-"*10 +" Exit the program.")
        

def check_status(account_detail):
    booking_key = 'Booking Status'
    if booking_key not in account_detail:
        account_detail[booking_key] = 'Not Booked'
        account_detail['Date'] = ''
        account_detail['Time'] = ''
        account_detail['Suburb'] = ''
        return account_detail
    else:
        return account_detail

def check_booking_status(account_detail):
    account_detail = check_status(account_detail)
    if account_detail['Booking Status'] == 'Not Booked':
        print(f"Booking Status : {account_detail['Booking Status']}")
        return account_detail
    else:
        print(f"Booking Status : {account_detail['Booking Status']}")
        print(f"Date : {account_detail['Date']}, Time : {account_detail['Time']}, suburb : {account_detail['Suburb']} ")
    return account_detail


def book_online_dkt(account_detail):
    account_detail = check_status(account_detail)
    if account_detail['Booking Status'] == 'Booked':
        print("********* You have already booked your DKT")
        return account_detail
    else:
        date = input("Enter the date for booking YYYY-MM-DD : ")
        time = input("Enter the time for booking HH:MM:SS : ")
        suburb = input("Enter suburb : ")
        account_detail['Date'] = date
        account_detail['Time'] = time
        account_detail['Suburb'] = suburb
        print("********* Your booking is successfully made for date,time, and suburb")
        account_detail['Booking Status'] = 'Booked'
        return account_detail

def reschedule(account_detail):
    account_detail = check_status(account_detail)
    if account_detail['Booking Status'] == 'Not Booked':
        print("********* You have no booking and go to book your test.")
        return account_detail
    else:
        print("********* Reschedule Booking")
        date = input("Enter the date for ticket booking YYYY-MM-DD : ")
        time = input("Enter the time for ticket booking HH:MM:SS : ")
        suburb = input("Enter suburb : ")
        account_detail['Date'] = date
        account_detail['Time'] = time
        account_detail['Suburb'] = suburb
        print("********* Your test successfully reschedule for data, time, and suburb.")
        return account_detail

def cancel_booking(account_detail):
    account_detail = check_status(account_detail)
    if account_detail['Booking Status'] == 'Not Booked':
        print("********* You don’t have any test booking")
        return account_detail
    else:
        account_detail['Booking Status'] = 'Not Booked'
        account_detail['Date'] = ''
        account_detail['Time'] = ''
        account_detail['Suburb'] = ''
        print("********* Your booking successfully cancelled for date, time, and suburb")
        return account_detail
def log_out(account_detail):
    print("xxxxxxxxxx The account has been logged out")
    print("xxxxxxxxxx Thank you for visiting Service NSW online")
    create_file(account_detail)
    return None

def exit_program(account_detail):
    create_file(account_detail)

    
is_program_started = True 
is_account_found = init_system(account_file) 
account_detail = None

while(is_program_started):
    if account_detail == False:
        break
    if(is_program_started == False):
        break
    if(is_account_found):
        account_detail = log_in(account_file)
    else:
        sign_up()
        account_detail = log_in(account_file)
        
    if account_detail == False:
        break
        
    if account_detail != None and account_detail != False:
        show_menu(1)
        
    while(account_detail != None and account_detail != False):
        choice = input("Choose an option: ")
        if choice == '1':
            show_menu(1)
        elif choice == '2':
            account_detail = check_booking_status(account_detail)
            show_menu(2)
        elif choice == '3':
            account_detail = book_online_dkt(account_detail)
            show_menu(2)
        elif choice == '4':
            account_detail = reschedule(account_detail)
            show_menu(2)
        elif choice == '5':
            account_detail = cancel_booking(account_detail)
            show_menu(2)
        elif choice == '0':
            account_detail = log_out(account_detail)
            show_menu(3)
            logout_option = input("Choose an option : ")
            if logout_option == '9':
                break
            elif logout_option == '00':
                choice = '00'
        elif choice == '00':
            exit_program(account_detail)
            is_program_started = False
            break

print("*** Exit Program ***")
```
